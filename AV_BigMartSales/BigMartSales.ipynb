```python
def combineData():
    train = pd.read_csv("datasets/train.csv")
    test = pd.read_csv("datasets/test.csv")
    target = train.Item_Outlet_Sales
    train.drop('Item_Outlet_Sales', 1, inplace=True)
    combined = train.append(test)
    combined.reset_index(inplace=True)
    combined.drop(['index','Outlet_Identifier','Item_Identifier'], inplace=True, axis=1)
    return combined
```

```python
CombinedData = combineData()
CombinedData
```

```python
def featureScaling(df):
#     df /= df.max()
#Min Max normalization
    df -= df.min()
    df /= (df.max()-df.min())
    return df
```

```python
CombinedData['Item_Visibility'] = featureScaling(CombinedData['Item_Visibility'])
```

```python
CombinedData.head(10)
```

```python
def recover_Train_Test_Target():
    global CombinedData , Train_Data
    targets = Train_Data.Item_Outlet_Sales
    train = CombinedData.head(8523)
    test = CombinedData.iloc[8523:]
    return train,targets,test 
    
```

```python
X,y,Test = recover_Train_Test_Target()
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train , y_test = train_test_split(X,y,test_size=0.1)
```

```python
# from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.linear_model import LinearRegression
```

```python
# model = XGBClassifier()
# model = CatBoostClassifier(iterations=100)
# model = LGBMClassifier()
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test,y_test)
accuracy
```

```python
y_pred = model.predict(X_test)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
rms = sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rms)
```
